# 2

```python
total_employees=df['Team'].value_counts().sum()
```

```python
total_employees
```

```python
employee_distribution = df['Team'].value_counts()
```

```python
employee_distribution
```

```python
percentage_split_of_employee = (employee_distribution)/(total_employees)*100
```

```python
percentage_split_of_employee
```

```python
employee_distribution.plot(kind='hist',color='b')
plt.xlabel('Number of players')
plt.ylabel('Teams')
plt.title('Players and their Team')
plt.show()
```

# 3

```python
employees = df['Position'].value_counts()
```

```python
employees
```

```python
employees.plot(kind='pie')
plt.legend(title = 'Players position',loc ='lower right')
plt.show()
```
